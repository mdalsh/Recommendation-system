```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_products = pd.read_csv('Extended_Products_Dataset__25_Products_ (1).csv')
df_invoices = pd.read_csv('Invoices_Dataset_for_Association_Rules.csv')
```

```python
print(df_products.isnull().sum())
```

```python
print(df_invoices.isnull().sum())
```

```python
df_products = df_products.dropna()
```

```python
numerical_cols = df_products.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    Q1 = df_products[col].quantile(0.25)
    Q3 = df_products[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_products.loc[:, col] = df_products[col].clip(lower=lower, upper=upper)

print(df_products.describe())
```

```python
transactions = df_invoices.groupby('InvoiceID')['ProductID'].apply(list).tolist()
te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df_trans, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
features_for_clustering = ['Price', 'Rating', 'Stock', 'WeightKg', 'VolumeCm3', 'PowerWatt', 'WarrantyYears']
df_cluster_data = df_products[features_for_clustering].copy()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cluster_data)
inertias = []
K_range = range(1, 10) 
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(scaled_features)
df_products['Cluster'] = cluster_labels
df_products.to_csv('products_with_clusters.csv', index=False)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_products, x='Price', y='Rating', hue='Cluster', palette='viridis', s=100)
plt.title('Clustering of Products (Price vs Rating)')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
cluster_summary = df_products.groupby('Cluster')[features_for_clustering].mean()
print(cluster_summary)
```

```python
features_for_clustering_no_price = ['Rating', 'Stock', 'WeightKg', 'VolumeCm3', 'PowerWatt', 'WarrantyYears']
df_cluster_data_no_price = df_products[features_for_clustering_no_price].copy()
scaler = StandardScaler()
scaled_features_no_price = scaler.fit_transform(df_cluster_data_no_price)
optimal_k = 3
kmeans_no_price = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels_no_price = kmeans_no_price.fit_predict(scaled_features_no_price)
df_products['Cluster_No_Price'] = cluster_labels_no_price
sns.set(style="whitegrid")
inertias = []
K_range = range(1, min(10, len(df_cluster_data_no_price) + 1)) 
for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(scaled_features_no_price)
    inertias.append(kmeans_temp.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o', linestyle='-', color='b')
plt.title('Elbow Method')
plt.xlabel('  (K)')
plt.ylabel(' (Inertia)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 7))
scatter = sns.scatterplot(
    data=df_products,
    x='Rating',
    y='Stock',
    hue='Cluster_No_Price',
    palette='Set1', 
    s=100, 
    alpha=0.8
)
plt.title('   (Rating vs Stock) -  ')
plt.xlabel(' (Rating)')
plt.ylabel('  (Stock)')
plt.legend(title=' (Cluster)')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_products, x='Cluster_No_Price', y='Price', palette='Set1')
plt.title('     (    )')
plt.xlabel('  (Cluster_No_Price)')
plt.ylabel(' (Price)')
plt.tight_layout()
plt.show()
# )   (Bar Plot)     
cluster_counts = df_products['Cluster_No_Price'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette='viridis')
plt.title('     ( )')
plt.xlabel('  (Cluster_No_Price)')
plt.ylabel(' ')
for i, v in enumerate(cluster_counts.values):
    plt.text(i, v + 0.2, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
